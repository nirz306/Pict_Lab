Feature Distribution


```python
sns.histplot(data=data, x='sepal_length',bins=10)
```

```python
sns.histplot(data=data, x='sepal_width', bins=10)
```

```python
sns.histplot(data=data, x='petal_length',bins=10)
```

3. Visualizing feature distribution With Box Plot


```python
sns.boxplot(data=data.drop(["species"],axis=1))
```

```python
sns.boxplot(data=data, x="species", y="sepal_length")
```

```python
sns.boxplot(data=data,x="species", y="sepal_width")
```

```python
sns.boxplot(data=data, x="species", y="petal_length")
```

```python
sns.boxplot(data=data, x="species",y="petal_width")
```
