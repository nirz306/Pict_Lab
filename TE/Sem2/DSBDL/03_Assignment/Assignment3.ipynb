```python
#operations on iris dataset 
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Id'])
```

```python
#üîç Summary statistics grouped by Species
grouped_stats = pd.DataFrame(df.groupby('Species').describe())
grouped_stats
```

```python
grouped_stats.mean()
```

```python
grouped_stats.median()
```

```python
# Create a list of petal lengths for each species
petal_lengths = grouped_stats['PetalLengthCm'].apply(list)

print("\nPetal Lengths for each species:")
print(petal_lengths)
```

```python
grouped_stats.head()
```

```python
grouped_stats = df.groupby('Species')
summary = grouped_stats.agg({
    'SepalLengthCm': ['mean', 'median', 'min', 'max', 'std'],
    'SepalWidthCm': ['mean', 'median', 'min', 'max', 'std'],
    'PetalLengthCm': ['mean', 'median', 'min', 'max', 'std'],
    'PetalWidthCm': ['mean', 'median', 'min', 'max', 'std']
})

print("\nCustom Summary Table:")
print(summary)
```
